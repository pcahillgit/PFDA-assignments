In this assignment I will load in a weather dataset and visualise some of the data in it.

The dataset in question provides data on various weather coditions at a particular location (Tuam, Co. Galway) over 15 minute intervals. I will use this dataset to plot temperature over time at this location.

Firstly, I will import the libraries:

```python
# Data frames
import pandas as pd

# Plotting
import matplotlib.pyplot as plt
```

Next, I will load the dataset from a URL and guage the max and min temps to inform the limits I might want to set on the Y axis:

```python
# Load the data set
df = pd.read_csv("https://raw.githubusercontent.com/andrewbeattycourseware/PFDA-courseware/refs/heads/main/assignment/weatherreadings1.csv")

max_temp = df['dryBulbTemperature_Celsius'].max()
min_temp = df['dryBulbTemperature_Celsius'].min()

print(max_temp)
print(min_temp)
```

Having done this I can now work with the data. I'm going to plot temperature over time:

```python
# Changing 'reportStartDateTime' to datetime
df['reportStartDateTime'] = pd.to_datetime(df['reportStartDateTime'])

# Creating figure and axes and plotting temperature as o's with connecting lines.
fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(df['reportStartDateTime'], df['dryBulbTemperature_Celsius'], marker='o', linestyle='-')

# Axis Labels and Y limit (not that the Tuam temperature goes that high very often!)
ax.set_xlabel('Date and Time (24HR)')
ax.set_ylabel('Temperature (°C)')
ax.set_ylim(0, 25)

# X labels at 45 degrees
plt.xticks(rotation=45)

# Filling colour under line
ax.fill_between(df['reportStartDateTime'], df['dryBulbTemperature_Celsius'], color='lightblue', alpha=0.5)

# Title
plt.title('Temperature in Tuam, Co. Galway (2nd-3rd October 2024)')

# Displaying the plot
plt.show()
```
